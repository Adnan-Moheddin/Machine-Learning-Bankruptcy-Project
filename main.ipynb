Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('dataFolder/data.csv')
data = pd.DataFrame(data)
data.head()
```

Split the dataset into features X and target variables y, and then split them into training and test sets using a 70/30 split, using a random seed of 43. 

Normalize Data using the MinMax Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X = data.iloc[:, 1:95]
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
y = data['Bankrupt?']

(X_train, X_test, Y_train, Y_test) = train_test_split(X, y, test_size = 0.3, random_state = 43)
print(Y_test.shape)
print(Y_train.shape)
```

Random Forests Implementation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score

parameters = {'n_estimators': [50, 100],
              'max_depth': [ 3, 4, 5],
              'max_features': [40, 50]
             }

randomForestClassifier = RandomForestClassifier()
rf = GridSearchCV(randomForestClassifier, parameters, scoring = 'f1')
randomForest = rf.fit(X_train, Y_train)
```

```python
Y_testReshape = Y_test.values.reshape(1, -1)
pred = randomForest.best_estimator_.predict(Y_testReshape)
randomForestF1 = f1_score(Y_testReshape, pred)
print(randomForestF1)
```
